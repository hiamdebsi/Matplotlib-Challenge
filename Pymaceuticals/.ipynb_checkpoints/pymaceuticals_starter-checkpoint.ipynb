## Observations and Insights 

```python
#Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from sklearn import datasets

#Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

#Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
#Checking the number of mice
mouse_metadata.count()
```

```python
#Checking the number of study results
study_results.count()
```

```python
#Drop any duplicates from both data sets to ensure having a clean data set before combining them
study_results_clean=study_results.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
mouse_metadata_clean=mouse_metadata.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
```

```python
#Checking each set of data post duplication removal
study_results_clean.count()
```

```python
mouse_metadata_clean.count()
```

```python
#Combine the data into a single dataset
merged_data_df=pd.merge(study_results_clean, mouse_metadata_clean, on='Mouse ID', how='inner')

#Display the data table for preview
merged_data_df
```

## Summary Statistics

```python
#Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
#Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
#mean, median, variance, standard deviation, and SEM of the tumor volume
#Assemble the resulting series into a single summary dataframe
print(f'Using the Groupby Method:')

groupby_drug_regimen_df=merged_data_df.groupby(['Drug Regimen'])
groupby_drug_regimen_df.count()
```

```python
#Create series for each statistical value required, and then convert them into a Dataframe
#MEAN
summary_mean = pd.Series(groupby_drug_regimen_df['Tumor Volume (mm3)'].mean(), name='mean')
summary_mean_df=pd.DataFrame(summary_mean)

#MEDIAN
summary_median = pd.Series(groupby_drug_regimen_df['Tumor Volume (mm3)'].median(), name='median')
summary_median_df=pd.DataFrame(summary_median)

#VARIANCE
summary_variance = pd.Series(groupby_drug_regimen_df['Tumor Volume (mm3)'].var(), name='variance')
summary_variance_df=pd.DataFrame(summary_variance)

#STD
summary_std = pd.Series(groupby_drug_regimen_df['Tumor Volume (mm3)'].std(), name='std')
summary_std_df=pd.DataFrame(summary_std)

#SEM
summary_sem = pd.Series(groupby_drug_regimen_df['Tumor Volume (mm3)'].sem(), name='sem')
summary_sem_df=pd.DataFrame(summary_sem)

#Concatenate all the values calculatated into one Dataframe
summary_all_df=pd.concat([summary_mean_df, summary_median_df, summary_variance_df, summary_std_df, summary_sem_df], axis=1)
summary_all_df
```

```python
#Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
#Using the aggregation method, produce the same summary statistics in a single line
print(f'Using the Aggregation Method:')

summary_all_agg =  merged_data_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_all_agg
```

## Quartiles, Outliers and Boxplots

```python
#Calculate the final tumor volume of each mouse across four of the treatment regimens:  
#Capomulin, Ramicane, Infubinol, and Ceftamin
#Start by getting the last (greatest) timepoint for each mouse
#Merge this group df with the original dataframe to get the tumor volume at the last timepoint
tumor_volume = merged_data_df.groupby(['Drug Regimen', 'Mouse ID']).last()[['Timepoint', 'Tumor Volume (mm3)']]
tumor_volume_df = tumor_volume.reset_index()
tumor_volume_df.head()
```

```python
#Put treatments into a list for for loop (and later for plot labels)
drug_regimen = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

#Create empty list to fill with tumor volume data (for plotting)
tumor_volume_list = []

#Create a for loop in order to run through all the calculations for each drug regimen
for drug in drug_regimen:
    
    #Calculate the IQR and quantitatively determine if there are any potential outliers. 
    #Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volume = tumor_volume_df.loc[tumor_volume_df['Drug Regimen'] == drug,'Tumor Volume (mm3)']

    #Add subset 
    tumor_volume_list.append(tumor_volume) 
    
    #Calculate the IQR
    quartiles =tumor_volume.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq

    #Lower and upper bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)

    #Determine outliers using upper and lower bounds
    outliers = tumor_volume.loc[(tumor_volume < lower_bound) & (tumor_volume > upper_bound)]

    #Print out results for each drug regimen
    print(f"{drug} Regimen")
    print(f"The lower quartile of the {drug} Regimen is: {lowerq}")
    print(f"The upper quartile of the {drug} Regimen is: {upperq}")
    print(f"The interquartile range of the {drug} Regimen is: {iqr}")
    print(f"The median of the {drug} Regimen is: {quartiles[0.5]} ")
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
    print('----------------------------------------------------------')
```

```python
#Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig3, ax3 = plt.subplots(figsize=(15, 10))

#Add required labels and format accordingly
ax3.set_title('Final Tumor Volume vs Drug Regimen',fontsize =25, fontweight='bold')
ax3.set_ylabel('Final Tumor Volume (mm3)',fontsize = 14)
ax3.set_xlabel('Drug Regimen',fontsize = 14)
ax3.boxplot(tumor_volume_list, labels=drug_regimen, widths = 0.4)

#Save plot as an image for easy reference
plt.savefig("../Images/Box_Plot.png")

#Print plot
plt.tight_layout()
plt.show()
```

## Line and Scatter Plots

```python
#Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
#Generate dataframe for the  Capomulin Regimen
Capomulin_df = merged_data_df.loc[merged_data_df['Drug Regimen'] == 'Capomulin']

#Identify specific mouse ID and get data of the mouse id. Will be using mouse m601.
mouse_id = 'm601'
mouse_df = Capomulin_df.loc[Capomulin_df['Mouse ID'] == mouse_id]

#Generate a line plot
fig4, ax4 = plt.subplots(figsize=(15, 10))
ax4 = plt.plot(mouse_df['Timepoint'], mouse_df['Tumor Volume (mm3)'], marker='o')

#Add required labels
plt.title(f"Mouse ID {mouse_id}'s Tumor Volume Progression with the Capomulin Regimen", fontsize =25, fontweight='bold')
plt.xlabel('Timepoint (Days)', fontsize = 14)
plt.ylabel('Tumor Volume (mm3)', fontsize = 14)

#Save plot as an image for easy reference
plt.savefig("../Images/m601_Line_Plot.png")

#Print plot
plt.tight_layout()
plt.show()
```

```python
#Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
#Calculate mean of Weight (g) and Tumor Volume (mm3) based on Mouse ID
groupby_mouseid = Capomulin_df.groupby('Mouse ID').mean()
mice_weight_avg = groupby_mouseid['Weight (g)']
mice_tumor_avg = groupby_mouseid['Tumor Volume (mm3)']

#Generate a scatter plot
fig5, ax5 = plt.subplots(figsize=(15, 10))
ax5 = plt.scatter(mice_weight_avg, mice_tumor_avg)

#Add required labels
plt.title('Average Weight vs Average Tumor Volume with the Capomulin Regimen', fontweight='bold', fontsize=25)
plt.xlabel('Average Weight (g)', fontsize=14)
plt.ylabel('Average Tumor Volume (mm3)', fontsize=14)

#Save plot as an image for easy reference
plt.savefig("../Images/Scatter_Plot.png")

#Print plot
plt.tight_layout()
plt.show()
```

## Correlation and Regression

```python
#Calculate the correlation coefficient and linear regression model 
#for mouse weight and average tumor volume for the Capomulin regimen

#Define the x and y values
x_values = mice_weight_avg
y_values = mice_tumor_avg

#Create the Line Regression, Line Equation, and r value
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
r_squared = round(rvalue**2,2)

#Create the scatter plot
fig6, ax6 = plt.subplots(figsize=(15, 10))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,37),fontsize=15,color="red")
plt.annotate(f'r-squared = {r_squared}', (20,36), fontsize=15, color="red")
plt.title('Average Weight vs Average Tumor Volume with the Capomulin Regimen', fontweight='bold', fontsize=25)
plt.xlabel('Average Weight (g)', fontsize=14)
plt.ylabel('Average Tumor Volume (mm3)', fontsize=14)

#Print r value
print(f"The r-squared is: {rvalue**2}")

#Print correlation coefficient
print(f"The correlation coefficient between the average mouse weight and average tumor volume is {round(st.pearsonr(x_values,y_values)[0],2)}")

#Save plot as an image for easy reference
plt.savefig("../Images/Correlation_Regression.png")

#Print plot
plt.show()
```
